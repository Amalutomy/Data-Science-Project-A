```python
import pandas as pd

# Define the SYNOPCode values and other parameters
synop_codes = [0, 4, 5, 6, 7, 8]
distance_ranges = [
    (2012, 2022),
    (2955, 2965),
    (4818, 4827)
]
time_range = (0, 19)

# Initialize an empty DataFrame to store results
downsample_data = pd.DataFrame()
all_rows_synop_code_3 = data[data['SYNOPCode'] == 3]
downsample_data = pd.concat([downsample_data, all_rows_synop_code_3])

# Loop through each SYNOPCode
for target_synop_code in synop_codes:
    # Filter the DataFrame based on the given ranges and SYNOPCode
    filtered_df = data[
        (data['SYNOPCode'] == target_synop_code) &
        (
            ((data['RDistance'] >= distance_ranges[0][0]) & (data['RDistance'] <= distance_ranges[0][1])) |
            ((data['RDistance'] >= distance_ranges[1][0]) & (data['RDistance'] <= distance_ranges[1][1])) |
            ((data['RDistance'] >= distance_ranges[2][0]) & (data['RDistance'] <= distance_ranges[2][1]))
        ) &
        (data['Time'] >= time_range[0]) & (data['Time'] <= time_range[1])
    ]

    # Randomly select 191 rows from the filtered DataFrame
    if len(filtered_df) >= 191:  # Ensure there are enough rows to sample
        selected_entries = filtered_df.sample(n=191, random_state=42)
    else:
        # If there are fewer than 191 rows, take all available rows
        selected_entries = filtered_df

    # Append the selected entries to the results DataFrame
    downsample_data = pd.concat([downsample_data, selected_entries])

# Display the result
downsample_data
```

```python
# Group by SYNOPCode and count the number of rows in each group
grouped_data = downsample_data.groupby('SYNOPCode').size()

# Display the number of rows in each group
print(grouped_data)
```

```python
downsample_data = downsample_data.drop('RDistance', axis=1)
```
