PLOTS

```python
import matplotlib.pyplot as plt
data_group_fso = {
    'SYNOPCode': [0, 3, 4, 5, 6, 7, 8],
    'RandomForest_RMSE': [2.28, 1.80, 1.00, 1.50, 1.58, 1.58, 1.44],
    'ExtraTrees_RMSE': [2.16, 1.33, 0.87, 1.36, 1.52, 1.28, 1.32]
}

df_group_fso = pd.DataFrame(data_group_fso)

# Plot for GROUP FSO
plt.figure(figsize=(10, 6))
plt.plot(df_group_fso['SYNOPCode'], df_group_fso['RandomForest_RMSE'], marker='o', label='Random Forest RMSE')
plt.plot(df_group_fso['SYNOPCode'], df_group_fso['ExtraTrees_RMSE'], marker='s', label='Extra Trees RMSE')
plt.title('GROUP FSO: Random Forest vs. Extra Trees Regressor RMSE')
plt.xlabel('SYNOPCode')
plt.ylabel('RMSE')
plt.legend()
plt.grid(True)
plt.show()

# Data for GROUP RFL Random Forest vs. Extra Trees Regressor
data_group_rfl = {
    'SYNOPCode': [0, 3, 4, 5, 6, 7, 8],
    'RandomForest_RMSE': [0.68, 0.40, 0.49, 0.68, 1.44, 0.23, 0.76],
    'ExtraTrees_RMSE': [0.64, 0.27, 0.48, 0.64, 1.03, 0.22, 0.72]
}

df_group_rfl = pd.DataFrame(data_group_rfl)

# Plot for GROUP RFL
plt.figure(figsize=(10, 6))
plt.plot(df_group_rfl['SYNOPCode'], df_group_rfl['RandomForest_RMSE'], marker='o', label='Random Forest RMSE')
plt.plot(df_group_rfl['SYNOPCode'], df_group_rfl['ExtraTrees_RMSE'], marker='s', label='Extra Trees RMSE')
plt.title('GROUP RFL: Random Forest vs. Extra Trees Regressor RMSE')
plt.xlabel('SYNOPCode')
plt.ylabel('RMSE')
plt.legend()
plt.grid(True)
plt.show()

# Data for Generic FSO Random Forest vs. Extra Trees Regressor
data_generic_fso = {
    'Model': ['Random Forest Regressor', 'Extra Trees Regressor'],
    'RMSE': [1.74, 1.56]
}

df_generic_fso = pd.DataFrame(data_generic_fso)

# Plot for Generic FSO
plt.figure(figsize=(8, 6))
plt.bar(df_generic_fso['Model'], df_generic_fso['RMSE'], color=['blue', 'green'])
plt.title('Generic FSO: Random Forest vs. Extra Trees Regressor RMSE')
plt.ylabel('RMSE')
plt.grid(axis='y')
plt.show()

# Data for Generic RFL Random Forest vs. Extra Trees Regressor
data_generic_rfl = {
    'Model': ['Random Forest Regressor', 'Extra Trees Regressor'],
    'RMSE': [0.86, 0.73]
}

df_generic_rfl = pd.DataFrame(data_generic_rfl)

# Plot for Generic RFL
plt.figure(figsize=(8, 6))
plt.bar(df_generic_rfl['Model'], df_generic_rfl['RMSE'], color=['blue', 'green'])
plt.title('Generic RFL: Random Forest vs. Extra Trees Regressor RMSE')
plt.ylabel('RMSE')
plt.grid(axis='y')
plt.show()
```

```python
# Group FSO Model Data
group_data = {
    'SYNOPCode': [0, 3, 4, 5, 6, 7, 8],
    'Optimal_RMSE': [2.16, 1.33, 0.87, 1.36, 1.52, 1.28, 1.32],
    'Optimal_R2': [0.72, 0.98, 0.95, 0.82, 0.79, 0.90, 0.79]
}

df_group = pd.DataFrame(group_data)

# Generic Model Data
generic_data = {
    'Model': ['Generic'],
    'RMSE': [0.73],
    'R2': [0.92]
}

df_generic = pd.DataFrame(generic_data)

# Plot RMSE Comparison
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.bar(df_group['SYNOPCode'], df_group['Optimal_RMSE'], color='blue', alpha=0.6, label='Group FSO RMSE')
plt.axhline(y=df_generic['RMSE'].values[0], color='orange', linestyle='--', label='Generic RMSE')
plt.xlabel('SYNOPCode')
plt.ylabel('RMSE')
plt.title('RMSE Comparison: Group FSO vs Generic Model Extra Trees Regressor')
plt.legend()

# Plot R Comparison
plt.subplot(1, 2, 2)
plt.bar(df_group['SYNOPCode'], df_group['Optimal_R2'], color='green', alpha=0.6, label='Group FSO R')
plt.axhline(y=df_generic['R2'].values[0], color='orange', linestyle='--', label='Generic R')
plt.xlabel('SYNOPCode')
plt.ylabel('R')
plt.title('R Comparison: Group FSO vs Generic Model')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

# Group RFL Model Data
group_data_rfl = {
    'SYNOPCode': [0, 3, 4, 5, 6, 7, 8],
    'Optimal_RMSE': [0.64, 0.27, 0.48, 0.64, 1.03, 0.22, 0.72],
    'Optimal_R2': [0.90, 0.99, 0.94, 0.94, 0.91, 0.94, 0.90]
}

df_group_rfl = pd.DataFrame(group_data_rfl)

# Generic RFL Model Data
generic_data_rfl = {
    'Model': ['Generic'],
    'RMSE': [0.73],
    'R2': [0.92]
}

df_generic_rfl = pd.DataFrame(generic_data_rfl)

# Plot RMSE Comparison for RFL
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.bar(df_group_rfl['SYNOPCode'], df_group_rfl['Optimal_RMSE'], color='blue', alpha=0.6, label='Group RFL RMSE')
plt.axhline(y=df_generic_rfl['RMSE'].values[0], color='orange', linestyle='--', label='Generic RFL RMSE')
plt.xlabel('SYNOPCode')
plt.ylabel('RMSE')
plt.title('RMSE Comparison: Group RFL vs Generic RFL Model')
plt.legend()

# Plot R Comparison for RFL
plt.subplot(1, 2, 2)
plt.bar(df_group_rfl['SYNOPCode'], df_group_rfl['Optimal_R2'], color='green', alpha=0.6, label='Group RFL R')
plt.axhline(y=df_generic_rfl['R2'].values[0], color='orange', linestyle='--', label='Generic RFL R')
plt.xlabel('SYNOPCode')
plt.ylabel('R')
plt.title('R Comparison: Group RFL vs Generic RFL Model')
plt.legend()

plt.tight_layout()
plt.show()
```
